```python
# Passo 1 - Percorrer todos os arquivos da pasta base de dados (pasta vendas)
import os #biblioteca para leitura de arquivos
import pandas as pd # pd significa abreviação de pandas

lista_arquivos = os.listdir("/content/drive/MyDrive/python/Vendas")
#print(lista_arquivos)
display(lista_arquivos)#funciona somente no colab
###__###
#lista_email = ['karina@gmail.com','kaique@gmail.com','steph@gmail.com']
#display(lista_email)
```

```python
#Passo 2 - Definir o caminho da bases de dados de vendas
# estrutura de repetição
#contar = 0
#while (contar<=5):
 # print(contar)
  #contar = contar + 1
  ###___###
for arquivo in lista_arquivos:
 # print(arquivo) # mostrando só o arquivo
  print(f"/content/drive/MyDrive/python/Vendas/{arquivo}")
#for numero in [1,2,3,4,5]: # outra estrutura de repetição
 #print(numero)
```

```python
#Passo 3 - Tratar / Compilar as bases de dados
#importar uma bilioteca de banco de dados
# estrutura de condição if
tabela_total = pd.DataFrame()
for arquivo in lista_arquivos:
  if 'vendas' in arquivo.lower():
    tabela = pd.read_csv(f"/content/drive/MyDrive/python/Vendas/{arquivo}")
    tabela_total = tabela_total.append(tabela)
display(tabela_total)
```

```python
 #Passo 4 - Calcular a quantidade vendida de cada produto
tabela_produtos = tabela_total.groupby('Produto').sum()
tabela_produtos = tabela_produtos[['Quantidade Vendida']].sort_values(by="Quantidade Vendida", ascending=False)
display(tabela_produtos)
```

```python
 #Passo 5 - Calcular o faturamento de cada produto
tabela_total['faturamento'] = tabela_total['Quantidade Vendida'] * tabela_total['Preco Unitario']
tabela_produtos = tabela_total.groupby('Produto').sum()
tabela_produtos = tabela_produtos[['faturamento']].sort_values(by="faturamento",ascending=False)
display(tabela_produtos)
```

```python
 #Passo 6 - Calcular o faturamento por loja - criar um gráfico/dashboard 
tabela_lojas = tabela_total.groupby('Loja').sum()
tabela_lojas = tabela_lojas[['faturamento']].sort_values(by="faturamento",ascending=False)
display(tabela_lojas)
```

```python
 #criar um gráfico/dashboard
import plotly.express as px
grafico = px.bar(tabela_lojas,x=tabela_lojas.index,y='faturamento')
grafico.show()
```
